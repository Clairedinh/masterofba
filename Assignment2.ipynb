```python
# Find rows with outliers in age in years
Q1 = df['age_in_years'].quantile(0.25)
Q3 = df['age_in_years'].quantile(0.75)
IQR = Q3 - Q1

print ("interquartile range of age: " , IQR)

# Define the bounds for identifying outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

print ('lower bound:', lower_bound)
print ('upper bound:', upper_bound)

outliers = df[(df['age_in_years'] < lower_bound) | (df['age_in_years'] > upper_bound)]

# Count the number of rows with outliers
outliers_count = outliers.shape[0]

# Display the number of outliers
print(f"Number of rows with outliers in the 'age_in_years' column: {outliers_count}")
```

```python
import numpy as np

# List of columns to check for outliers
columns = ['paw_size', 'head_length', 'ear_size', 'skull_width', 'eye_diameter', 'chest_circumference', 'belly_circumference']

def replace_outliers_with_median(df, col):
    # Calculate Q1 (25th percentile) and Q3 (75th percentile) for each column
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    
    # Define lower and upper bounds
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    # Calculate the median of the column
    median = df[col].median()
    
    # Replace outliers with the median
    df[col] = np.where((df[col] < lower_bound) | (df[col] > upper_bound), median, df[col])

# Apply the function to each column, replacing outliers with the median
for col in columns:
    replace_outliers_with_median(df, col)

# Check the results
for col in columns:
    print(f"Unique values in {col} after replacement:", df[col].unique())
```

## 3.1. Is the mean head length of the Koalas significantly different from 92.0 mm?

```python
from statsmodels.stats.weightstats import ztest

# Hypothesized population mean
population_mean = 0.92

# Perform a one-sample z-test
z_statistic, p_value = ztest(df['head_length'], value=population_mean)

# Significance level
alpha = 0.05

# Step 4: Decide based on the p-value
if p_value < alpha:
    print("Reject the null hypothesis. The mean head length of the Koalas is significantly different from 92.00 mm.")
else:
    print("Fail to reject the null hypothesis. The data does not provide sufficient evidence that the mean head length of the Koalas is significantly different from 92.00 mm.")

# Print the z-statistic and p-value
print(f"statistic: {z_statistic:.4f}")
print(f"p_value: {p_value:.4f}")
```

## 3.2. Do male and female Koalas have significantly different mean head lengths?

```python
# Import the necessary module
from scipy import stats

# Separate the data into male and female groups
male_head_lengths = df[df['gender'] == 'male']['head_length']

female_head_lengths = df[df['gender'] == 'female']['head_length']

# Conduct an independent two-sample t-test
t_statistic, p_value = stats.ttest_ind(male_head_lengths, female_head_lengths)

# Print the results

print(f"T-statistic: {t_statistic:.4f}")
print(f"P-value: {p_value:.4f}")

# Determine significance
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: Male and female koalas have significantly different mean head lengths.")
else:
    print("Fail to reject the null hypothesis: Male and female koalas do not have significantly different mean head lengths.")
```

## 3.4. Can we predict the total length of a Koala based on multiple factors such as head length, skull width, and foot length?

```python
import statsmodels.api as sm
from statsmodels.stats.outliers_influence import variance_inflation_factor


# Define the independent variables (X) and dependent variable (y)
X = df[['head_length', 'skull_width', 'foot_length']]  # Independent variables
y = df['total_length']   # Dependent variable
```

```python
# Add a constant term to the model (for the intercept)
#sm.add_constant(X) adds a column of ones to the matrix X.
#The added column represents the constant term (also known as the intercept)
X = sm.add_constant(X)

# Calculate VIF for each feature
vif_data = pd.DataFrame()
vif_data['Feature'] = X.columns

#Contain two columns, the first is the name of the feature, second is VIF

#X.shape[1] tells us how many features (or columns) we have in our dataset X. If we have, say, 5 features, X.shape[1] would be 5.
#range(X.shape[1]): This creates a sequence of numbers from 0 to the number of features minus one (so in our example, from 0 to 4). Each number corresponds to a different feature.
#variance_inflation_factor(X.values, i): For each feature (represented by i), this function calculates the VIF. 
#The function takes in all the data from X (as a NumPy array using X.values) and the specific feature index i to calculate the VIF.
vif_data['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]

# Display the VIF data
print(vif_data)
```

```python
# Fit the multiple regression model. Do not specify the formula
model = sm.OLS(y, X).fit()

# Display the summary of the regression model
print(model.summary())
```

## 3.5. Do environmental factors such as state affect Koala’s physical characteristics?

```python
import pandas as pd
import statsmodels.api as sm
from statsmodels.formula.api import ols

# Assuming df is already defined
# One-hot encoding the 'habitat' column and dropping the first column
df = pd.get_dummies(df, columns=['habitat'], dtype = int, drop_first=True)


# List of physical characteristics to test
physical_characteristics = ['head_length', 'paw_size', 'ear_size', 'foot_length', 'total_length',
                            'skull_width', 'eye_diameter', 'chest_circumference', 'belly_circumference']

# Independent variable is one of the new columns, e.g., 'habitat_VIC'
independent_var = 'habitat_VIC'  # Make sure this column exists after encoding

# Loop through each characteristic and build a separate OLS model
for characteristic in physical_characteristics:
    # Create the formula using the characteristic and the independent variable
    ols_formula = f"{characteristic} ~ {independent_var}"
    
    # Build and fit the OLS model
    OLS = ols(formula=ols_formula, data=df)
    model = OLS.fit()

    # Print the summary for each characteristic
    print(f"Regression results for {characteristic} (dependent) vs {independent_var} (independent):")
    print(model.summary())
    print("\n" + "="*60 + "\n")
```

From the respective single regression model, I keep focus on the p-value of each corresponding result. 
p_value of regression model of state with head_length, total_length, skull_width, eye_diameter and belly_diameter all over 0.05, which means that each of them is not significant. Therefore, state does not significantly affect some koala's physical characteristics, including head_length, total_length, skull_width, eye_diameter, and belly_circumference.
