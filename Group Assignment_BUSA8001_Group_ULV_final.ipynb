**Evaluation Criteria**

The Mean Absolute Percentage Error (MAPE) is chosen to evaluate the model's performance. It measures the accuracy of predictions by calculating the average percentage difference between predicted and actual values. Unlike other metrics, MAPE is easier to interpret, particularly for non-technical stakeholders, as it indicates how much the predictions deviate from actual values in relative terms. A lower MAPE indicates better model accuracy.

**Formula**

$$
\text{MAPE} = \frac{1}{n} \sum_{i=1}^n \left| \frac{y_i - \hat{y}_i}{y_i} \right|\times 100
$$


**MAPE Calculation Steps:**

- Calculate Absolute Errors: Measure the absolute difference between each predicted value and the actual value.
- Divide by Actual Values: For each data point, divide the absolute error by the actual observed value to get the percentage error.
- Average the Percentage Errors: Compute the mean of all percentage errors.
- Multiply by 100: Multiply the mean by 100 to express MAPE as a percentage of the actual values.

**Task 2, Question 3**: Impute the missing values for all features in both the training and test datasets.   
(2 marks)

```python
def impute_missing_val(df):
    # Find columns with missing values and save their names into a list
    columns_with_missing_values = df.columns[df.isnull().any()].tolist()

    # Group by 'make_name' and 'model_name', then fill missing values with the same value within each group
    for column in columns_with_missing_values:
        if column in df.columns:  # Ensure the column exists
            df[column] = df.groupby(['make_name', 'model_name'])[column].transform(
                lambda x: x.fillna(method='ffill').fillna(method='bfill')
            )

    # Separate categorical and numerical columns
    categorical = ['body_type', 'city', 'engine_type', 'exterior_color', 'fuel_type', 'interior_color', 
                   'listing_color', 'make_name', 'model_name', 'transmission', 'wheel_system', 
                   'transmission_display','num_gears']
    boolean = ['franchise_dealer', 'is_new']
    numerical = ['back_legroom', 'city_fuel_economy', 'daysonmarket', 'dealer_zip', 'engine_displacement', 
                 'front_legroom', 'fuel_tank_volume', 'height', 'highway_fuel_economy', 'horsepower', 
                 'latitude', 'length', 'longitude', 'maximum_seating', 'mileage', 'savings_amount', 
                 'seller_rating', 'wheelbase', 'width', 'year', 'power_value', 'power_rpm', 
                 'torque_value', 'torque_rpm']


    # Impute missing values for categorical columns
    for column in categorical:
        if column in df.columns:  # Check if the column exists
            df[column].fillna(df[column].mode()[0], inplace=True)
    
    # Impute missing values for boolean columns
    for column in boolean:
        if column in df.columns:  # Check if the column exists
            df[column].fillna(df[column].mode()[0], inplace=True)

    # Impute missing values for numerical columns
    for column in numerical:
        if column in df.columns:  # Check if the column exists
            df[column].fillna(df[column].median(), inplace=True)

    return df

# Apply the function to train_df and test_df separately
train_df = impute_missing_val(train_df)
test_df = impute_missing_val(test_df)
```

**Explanation**

Handling missing data is crucial to maintaining the integrity and reliability of machine learning models. The chosen imputation strategies are designed to mitigate biases and errors, ensuring that the model performs optimally and accurately reflects the underlying data patterns.

- Forward and Backward Fill: Missing values in the dataset were imputed using forward-fill and backward-fill strategies within each group defined by make_name and model_name. This approach leverages the existing data patterns, allowing for logical consistency based on similar vehicle characteristics. Forward-fill uses the last known value to fill subsequent missing entries, while backward-fill uses the next known value, ensuring that the imputation remains contextually relevant and preserves relationships within the data.

- **Mode for Categorical Columns**: For categorical features, missing values were filled using the mode, or the most frequent value. This method is appropriate because it retains the most common attribute within each category, which minimizes the distortion of the data distribution. Using mode also prevents the introduction of noise, particularly in datasets where certain categories may have few occurrences.

- **Mode for Boolean Columns**: For boolean features, missing values were filled using the mode. This approach effectively retains the most frequently occurring value (True or False) within the boolean category, thereby minimizing the impact on the overall data distribution. Employing mode for boolean columns ensures that the imputation is based on existing patterns in the data, which aids in preserving the logical relationships and consistency necessary for accurate model predictions.

- **Median for Numerical Columns**: Missing values in numerical features were imputed with the median. The median is a robust measure of central tendency that is less sensitive to outliers compared to the mean. By employing the median, the imputation reflects a typical value within the dataset, which is especially important in datasets containing diverse car attributes. This approach enhances the reliability of the numerical features, contributing to a more stable model performance.

The function impute_missing_val() effectively applies these strategies to ensure that both the training and test datasets are complete, thereby enhancing the overall quality of the data used for model training and evaluation.

**Task 2, Question 4**: Encode all categorical variables appropriately as discussed in class. 

- Where multiple values are given for an observation encode the observation as 'other'. 
- Where a categorical feature contains more than 5 unique values, map the features into 5 most frequent values + 'other' and then encode appropriately. For instance, if appropriate, you could all map colours into {red, yellow, green, blue, purple, other} and then encode.  
(2 marks)

```python
# Store the VIN data
vin_column = test_df['vin'].copy()

import pandas as pd

# Define function to encode all categorical variables
def encode(df):
    # Remove sub-string in engine_type (if applicable)
    if 'engine_type' in df.columns:
        df['engine_type'] = df['engine_type'].str.split().str[0]
    
    # Identify categorical columns
    categorical_cols = df.select_dtypes(include=['object']).columns
    
    # If a column has more than 5 unique values, keep only the top 5 most frequent ones
    for col in categorical_cols:
        if df[col].nunique() > 5:
            top_5 = df[col].value_counts().head(5).index
            df[col] = df[col].where(df[col].isin(top_5), 'other')

    # One-hot encode the transformed categorical features
    encode_cols = ['body_type', 'engine_type', 'fuel_type', 'listing_color',
                   'make_name', 'model_name', 'transmission', 'wheel_system', 'transmission_display', 'num_gears']
    
    # Ensure we only encode columns that exist in the DataFrame
    encode_cols = [col for col in encode_cols if col in df.columns]
    
    # Perform one-hot encoding
    df = pd.get_dummies(df, columns=encode_cols, drop_first=True)  # drop_first to avoid multicollinearity
    
    # Identify the newly encoded columns
    encoded_columns = [col for col in df.columns if any(prefix in col for prefix in encode_cols)]
    
    # Ensure that the encoded columns are in integer format (0s and 1s)
    df[encoded_columns] = df[encoded_columns].astype(int)

    return df

# Apply function to train_df and test_df
train_df = encode(train_df)
test_df = encode(test_df)
```

**Explanation**

The objective of encoding categorical variables was to streamline them for machine learning models, ensuring that they contribute effectively to model performance. Categorical variables in the dataset varied significantly in their characteristics, with some features containing a high number of unique values, while others were limited to a few categories.

- **Characteristics of Categorical Variables**: Before encoding, categorical variables such as body_type, engine_type, and fuel_type exhibited varying distributions. For instance, body_type might have included several styles, while engine_type could encompass various specifications with unnecessary substrings that added complexity without contributing meaningfully to the analysis. Features with more than five unique values were particularly challenging, as rare categories could introduce noise and lead to overfitting in the model.

- **Simplifying Categorical Features**: To address these complexities, categorical features were first simplified by removing unnecessary substrings (e.g., from the engine_type column). For features with more than five unique values, only the top five most frequent categories were retained, while less common categories were grouped and encoded as 'other.' This approach allowed for the preservation of the most representative trends in the data, enhancing the model’s ability to generalize.

- **One-Hot Encoding**: After simplifying the categorical variables, one-hot encoding was applied to selected features. This process converts categorical values into a numerical format, enabling their use in machine learning algorithms. To mitigate the risk of multicollinearity—which can negatively impact model performance—dropping the first category of the one-hot encoded variables was implemented. This change ensures that the model does not suffer from redundancy among features, allowing for a clearer understanding of the relationships within the data.

By focusing on the most frequent categories and encoding them appropriately, the risk of overfitting due to rare categories is reduced, promoting a more robust model. The decision to drop the first column during one-hot encoding enhances interpretability and model performance by preventing redundancy. Overall, these encoding strategies ensure that the categorical variables are effectively transformed into a format conducive to accurate predictions.

* exterior_color and interior_color: These describe the car’s appearance, which is already captured by the listing_color column. Removing these prevents redundancy in the dataset.
* city: While this represents the vehicle’s location, the dealer_zip column provides more precise geographical information, making city unnecessary.
* franchise_dealer and is_new: These variables have minimal impact on price prediction, as other features, such as mileage, engine_displacement, horsepower, and fuel_type, provide a more comprehensive reflection of a vehicle's condition and value
* listed_date: This feature does not directly affect pricing and is better suited for analyses focused on market trends rather than model prediction.

By dropping these columns, we reduce dataset noise, improve interpretability, and simplify the model without losing essential information. This targeted feature selection helps enhance the model's ability to generalize to new data by focusing on relevant variables that contribute meaningfully to price prediction.

```python
# 2. Standardizing numerical columns

from sklearn.preprocessing import StandardScaler

# Define the numerical columns explicitly
numerical_cols = [
    'back_legroom', 'city_fuel_economy', 'daysonmarket', 
    'dealer_zip', 'engine_displacement', 'front_legroom', 
    'height', 'highway_fuel_economy', 'horsepower', 
    'latitude', 'longitude', 'maximum_seating', 
    'mileage', 'savings_amount', 'seller_rating', 
    'wheelbase', 'width', 'year', 'power_value', 
    'power_rpm', 'torque_value', 'torque_rpm'
]

# Initialize the StandardScaler
scaler = StandardScaler()

# Standardize numerical features in the training dataset
train_df[numerical_cols] = scaler.fit_transform(train_df[numerical_cols])

# Standardize numerical features in the test dataset using the same scaler
test_df[numerical_cols] = scaler.transform(test_df[numerical_cols])
```

**Explanation**

Standardizing numerical features is a critical preprocessing step that ensures all variables are on the same scale. This process is particularly important for machine learning algorithms sensitive to the scale of the data, such as Support Vector Machines (SVM), K-Nearest Neighbors (KNN), and models that rely on gradient descent optimization, including Linear Regression and eXtreme Gradient Boosting (XGBoost). Without standardization, features with larger ranges could dominate those with smaller ranges, leading to biased model performance and hindering the model’s ability to learn effectively from the data.

In this analysis, numerical features were standardized using the StandardScaler, which transforms the data to have a mean of 0 and a standard deviation of 1. This transformation was applied to the training dataset to fit the scaler, and the same scaling parameters were then used to standardize the test dataset. By ensuring that all numerical features are normalized, a level playing field is created for the model, enhancing convergence rates and overall predictive performance.

Additionally, for tree-based models like Random Forest, while they are generally less sensitive to feature scaling, standardizing numerical features can still contribute to more stable training and improved interpretability of feature importance. Establishing a consistent scale across features lays a solid foundation for the predictive modeling process that follows, allowing models to learn from the data more effectively.

## 3. Train the models and tune their hyperparameters via cross-validation. Discuss the fitted weights, optimised hyperparameter values, and their training dataset predictive performance.

To predict the price of used cars, 3 models will be applied: KNN regressor, random forest and XGBoost. To start, training dataset (train_df) will be split into training and validation subsets, with 70% will be used for training and 30% will be used for validation.

```python
# Data Preparation
from sklearn.model_selection import train_test_split

# Separate features and target from the training dataset
X_train = train_df.drop('price', axis=1)  # Use only the specified feature columns
y_train = train_df['price'] # Fixed typo from train_d to train_df

# Split the data into 70% train and 30% validation datasets
X_train_split, X_val_split, y_train_split, y_val_split = train_test_split(X_train, y_train, test_size=0.3, random_state=1)

# Display the shapes of the resulting datasets
X_train_split.shape, X_val_split.shape
```

### Model 1. KNN regressor model 

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_absolute_percentage_error

# Fit the KNN Regression model
knn_reg = KNeighborsRegressor()
knn_reg.fit(X_train_split, y_train_split)

# Predict on the validation split
knn_reg_preds = knn_reg.predict(X_val_split)

# Compute MAPE
knn_reg_mape = mean_absolute_percentage_error(y_val_split, knn_reg_preds)
print('knn_reg_mape:', knn_reg_mape)
```

```python
from sklearn.model_selection import GridSearchCV

# Define a parameter grid for KNN
param_grid = {
    'n_neighbors': [5, 10],  
    'weights': ['uniform', 'distance'],
    'p': [1, 2]           
}

# Initialize the KNeighborsRegressor
knn_model = KNeighborsRegressor()

# Initialize GridSearchCV 
grid_search = GridSearchCV(estimator=knn_model, 
                           param_grid=param_grid, 
                           scoring='neg_mean_absolute_percentage_error', 
                           cv=10,               
                           n_jobs=-1,           # Utilize all processors
                           verbose=2)           # Verbose output to show progress

# Fit the GridSearchCV to the training data
grid_search.fit(X_train_split, y_train_split)

# Print the best parameters found by GridSearchCV
print("Best parameters for KNN:", grid_search.best_params_)

# Use the best model to make predictions on validation data
best_knn_model = grid_search.best_estimator_
knn_preds = best_knn_model.predict(X_val_split)

# Compute MAPE on validation data using the best model
knn_mape = mean_absolute_percentage_error(y_val_split, knn_preds)
print('Best KNN MAPE:', knn_mape)
```

**Comment on model's performance**

The MAPE of the KNN regressor at approximately 0.1788 (or 17.9%) indicates a moderate level of accuracy in predicting the target variable, which means the model's predictions deviate 17.9% from the actual values. While a MAPE of 0.1788 shows reasonable predictive performance, there remains room for improvement to optimize the model further and achieve better results.

**Optimised Hyperparameters**

- n_neighbors: 5: This indicates that the model considers the 5 nearest neighbors when making predictions. The choice of 5 suggests a balance that allows the model to leverage local data without being overly influenced by distant points.

- p: 1: This parameter specifies the distance metric used to calculate the similarity between data points. Setting p=1 corresponds to the Manhattan distance, which sums the absolute differences between coordinates. This distance metric is particularly better when dealing with high-dimensional data.

- weights: 'distance': This parameter controls how each neighbor contributes to the prediction. By assigning weights based on the inverse of the distance to the neighbors, closer neighbors contribute more to the prediction than those further away. This is beneficial in scenarios where proximity implies greater relevance, thus enhancing the model's sensitivity to local patterns.

**Cross-Validation**
- Using 10 folds validation in GridSearchCV for tuning the KNN model allows for a thorough evaluation of the model’s performance across different hyperparameter settings. This method helps ensure that the selected hyperparameters not only fit the training data well but also generalize effectively to unseen data, ultimately leading to a more robust and reliable model.


### Model 2. Random forest

```python
from sklearn.ensemble import RandomForestRegressor

# Fit the Random Forest Regression model and predict on the validation split
rf_reg = RandomForestRegressor()
rf_reg.fit(X_train_split, y_train_split)
rf_reg_preds = rf_reg.predict(X_val_split)

# Compute MAPE
rf_reg_mape = mean_absolute_percentage_error(y_val_split, rf_reg_preds)
print('rf_reg_mape:',rf_reg_mape)
```

```python
# Define an expanded parameter grid to explore a broader range
param_grid = {
    'n_estimators': [100, 200, 300],  
    'max_depth': [10, 20, 30],        
    'min_samples_split': [2, 5],      
    'min_samples_leaf': [1, 2, 4]    
}

# Initialize the RandomForestRegressor
rf_reg = RandomForestRegressor(random_state=1)

# Initialize GridSearchCV with expanded parameters and CV
grid_search = GridSearchCV(
    estimator=rf_reg, 
    param_grid=param_grid, 
    scoring='neg_mean_absolute_percentage_error', 
    cv=10,                
    n_jobs=-1,           
    verbose=2            
)

# Fit the GridSearchCV to the training data
grid_search.fit(X_train_split, y_train_split)

# Print the best parameters found by GridSearchCV
print("Best parameters found: ", grid_search.best_params_)

# Use the best model to make predictions
best_rf_reg = grid_search.best_estimator_
rf_reg_preds = best_rf_reg.predict(X_val_split)

# Compute MAPE on validation data using the best model
rf_reg_mape = mean_absolute_percentage_error(y_val_split, rf_reg_preds)
print('Best rf_reg_mape:', rf_reg_mape)
```

**Comment on model's performance**

For the Random Forest model, the best MAPE acchieved by model during tuning is 0.1253, or about 12.5%, indicates a notable improvement over the KNN regressor model's MAPE of 17.9%. This lower MAPE suggests the model is more accurate in predicting used car prices, making it better suited for this task. The MAPE value of around 13% indicates the model's capability to generalize quite well to unseen data, making it a valuable tool for predicting used car prices.

**Optimised Hyperparameters**
- max_depth: 20: This parameter defines the maximum depth of the trees in the forest. A maximum depth of 20 indicates that the trees can become relatively deep, allowing them to capture more complex relationships in the data. This helps balance model complexity with the prevention of overfitting.

- min_samples_leaf: 1: The minimum number of samples required to be at a leaf node (end node). Each leaf node must contain at least one sample. By allowing leaf nodes to have as few as one sample, the model can capture very detailed patterns in the data, potentially leading to more accurate predictions.

- min_samples_split: 2:  This parameter sets the minimum number of samples required to split an internal node. With the result, a node requires at least two samples to be split, which is typical for building fully grown trees.

- n_estimators: 300: This parameter represents the number of trees in the forest. The forest consists of 300 trees, which helps improve the model’s accuracy by reducing variance in predictions.


**Cross-Validation**

The grid search tested 54 combinations of hyperparameters (from param_grid), and each combination was evaluated using 10-fold cross-validation. This means the model was trained and validated 10 times for each parameter set, resulting in 540 total fits. This process helps validate the model’s stability and generalization, making it more reliable when applied to unseen data.


### Model 3. eXtreme Gradient Boosting model

```python
import xgboost as xgb

# Initialize the XGBoost regression model and fit the training data
xgb_model = xgb.XGBRegressor(objective='reg:squarederror')
xgb_model.fit(X_train_split, y_train_split)

# Predict on the validation data
xgb_pred = xgb_model.predict(X_val_split)

# Calculate MAPE
xgb_mape = mean_absolute_percentage_error(y_val_split, xgb_pred)
# Print MAPE as a percentage
print('xgb_mape',xgb_mape)
```

```python
# Define a smaller parameter grid to optimize key XGBoost hyperparameters
param_grid = {
    'n_estimators': [100, 200],                   # Number of boosting rounds
    'learning_rate': [0.01, 0.05, 0.1],           # Learning rate (step size)
    'max_depth': [3, 6, 9],                       # Maximum depth of trees
    'subsample': [0.7, 1],                        # Fraction of samples used per tree
    'colsample_bytree': [0.7, 1],                 # Fraction of features used per tree
}

# Initialize the XGBoost Regressor with squared error as the objective
xgb_model = xgb.XGBRegressor(objective='reg:squarederror', random_state=1)

# GridSearchCV to tune the model
grid_search = GridSearchCV(estimator=xgb_model, 
                           param_grid=param_grid, 
                           scoring='neg_mean_absolute_percentage_error', 
                           cv=10,                # Reduced folds to make it less computationally expensive
                           n_jobs=-1,           # Use all processors
                           verbose=1)           # Print progress

# Fit the model on the training data
grid_search.fit(X_train_split, y_train_split)

# Best parameters and model after tuning
best_params = grid_search.best_params_
print("Optimized hyperparameters:", best_params)

# Predict on validation set using the best model
best_xgb_model = grid_search.best_estimator_
xgb_pred = best_xgb_model.predict(X_val_split)

# Calculate MAPE for validation predictions
xgb_mape = mean_absolute_percentage_error(y_val_split, xgb_pred)
print('Optimized xgb_mape:', xgb_mape)
```

**Comment on the model's performance**

The optimized XGBoost model achieved a MAPE of 11.73%, indicating that, on average, its predictions deviate by around 11.73% from actual car prices, marking it as the best-performing model so far. This number suggests the XGBoost model performs quite well in predicting used car prices.

- learning_rate: 0.05: The learning rate reduces the contribution of each tree to the final model. A lower learning rate, such as 0.05, indicates that the model is making more cautious updates, which can enhance the stability of the learning process and often leads to better generalization. However, this usually requires more trees (n_estimators) to achieve similar performance, which is addressed by setting n_estimators to 200.
- max_depth: 9: This parameter defines the maximum depth of the individual trees in the model. A depth of 9 allows the model to learn complex interactions in the data while helping to prevent overfitting by limiting how deeply the trees can grow.
- n_estimators: 200: This indicates that the model will use 200 trees in the ensemble. With the lower learning rate, this number is adequate for capturing the underlying patterns in the data while balancing computational efficiency.
